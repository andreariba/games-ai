```python
class AZLoss(tf.keras.losses.Loss):
    
    def __init__(self):
        super().__init__('none')
        self.p_loss_fn = tf.keras.losses.CategoricalCrossentropy()
        self.v_loss_fn = tf.keras.losses.MeanSquaredError()

    def call(self, y_true, y_pred):
        
        estimated_pis, estimated_vs = y_true
        predicted_pis, predicted_vs = y_pred
        
        p_losses = self.p_loss_fn(estimated_pis, predicted_pis)
        print(p_losses)
        v_losses = self.v_loss_fn(predicted_vs, estimated_vs)
        print(v_losses)
        
        return p_losses + v_losses
```

# Test Model

```python
import numpy as np
import tensorflow as tf
```

```python
tf.one_hot(tf.cast(np.array([[0., 0., -1., 0., 0., 0., 0., -1., 1.]])+1, tf.int32),3)
```

```python
def create_az_model():
    
    inputs = tf.keras.layers.Input((9))

    # One-Hot encoding
    x = tf.one_hot(tf.cast(inputs+1, tf.int32),3)

    # Residual tower NN
    x = tf.reshape(x, (-1,3,3,3))
    x = tf.keras.layers.Conv2D(filters=3**6, kernel_size=(3,3))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = tf.keras.layers.Conv2D(filters=3**4, kernel_size=1)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Flatten()(x)
    flatten_inputs = tf.keras.layers.Flatten()(inputs)
    x = tf.concat([flatten_inputs, x], axis=1)
    residual_tower = tf.keras.layers.ReLU()(x)

    # Policy NN
    x = tf.keras.layers.Conv1D(filters=2, kernel_size=1)(residual_tower[...,np.newaxis])
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = tf.keras.layers.Flatten()(x)
    p = tf.keras.layers.Dense(9, activation=tf.keras.activations.softmax)(x)

    # Value NN
    x = tf.keras.layers.Conv1D(filters=1, kernel_size=1)(residual_tower[...,np.newaxis])
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = tf.keras.layers.Flatten()(x)
    x = tf.keras.layers.Dense(256, activation=tf.keras.activations.relu)(x)
    v = tf.keras.layers.Dense(1, activation=tf.nn.tanh)(x)

    outputs = p, v

    model = tf.keras.Model(inputs=inputs, outputs=outputs, name="tictactoe_model")
    
    return model
```

```python
model = create_az_model()
tf.keras.utils.plot_model(model, show_shapes=True)
```
