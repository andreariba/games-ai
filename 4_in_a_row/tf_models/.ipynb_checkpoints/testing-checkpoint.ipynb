```python
import multiprocess as mp
from multiprocess.managers import BaseManager, SyncManager
```

```python
n_cores = mp.cpu_count()
```

```python
n_cores
```

```python
__name__
```

```python
def parallel_fn(job_n, cache):
    
    cache['job_b'] = job_n
    
    return job_n
```

```python
if __name__=="__main__":
    ctx = mp.get_context("spawn")

    with SyncManager() as manager:

        shared_cache = manager.dict()
        
        args = list(zip(range(n_cores), shared_cache))

        with mp.Pool(n_cores) as pool:

            result = pool.starmap(parallel_fn, args)

            print(result)
        
        print(shared_cache)
```

```python
if __name__=="__main__":
    
    args = list(zip(range(n_cores), {}))
    
    ctx = mp.get_context("spawn")

    pool = ctx.Pool(n_cores)

    result = pool.starmap(parallel_fn, args)

    print(result)
```
